#### <font color = "red">P5.- ¿Tipos de datos estadísticos?</font>



**1) Categoricos o cualitaitivos**: representan características y pueden tomar valores numéricos. Se clasifican en:

*1.1. Datos nominales*: representan unidades discretas y se usan para etiquetar variables que no tienen un valor cuantitativo. Estos datos no tienen un orden, aunque cambiara el orden de sus valores, no cambia su significado.

*1.2. Datos ordinales:* representan unidades discretas y ordenadas, pero su orden es importante. Se usan para medir características no numéricas.

**2) Numericos o cuantitativos:** se refieren a una medida o recuento. Se clasifican en:

*2.1. Datos discretos:* sus valores son distintos y separado; es decir, cuando los datos sólo pueden tomar ciertos valores. Este tipo de datos no se puede medir, pero se pueden contar; representa información que se puede clasificar.

*2.2. Datos continuos:* representan mediciones y, por lo tanto, sus valores no se pueden contar pero se pueden medir. Se clasifican en:

2.2.1. Datos de intervalo: Los datos de intervalo representan unidades ordenadas que tienen la misma diferencia. Es decir, cuando tenemos una variable que contiene valores numéricos que están ordenados y donde conocemos las diferencias exactas entre los valores. El problema es que podemos sumar y restar, pero no multiplicar, dividir o calcular razones. 

2.2.2. Datos de relación: son unidades ordenadas que tienen la misma diferencia. Los datos de relación son los mismos que los valores de intervalo, con la diferencia de que tienen un cero absoluto.

## Resultados Mundiales

Para poder seleccionar el conjunto de datos que nos interesa, por el momento podemos observar el formato de fechas que se está manejando en el conjunto de datos, que es el siguiente: Mes / Dia / Año

Por lo que, podemos auxiliarnos de la función loc para delimitar que columnas deberíamos seleccionar 

```python
confirmed_cases = confirmed.copy()
death_cases = death.copy()
```

```python
cols = confirmed_cases.keys()
confirmed = confirmed_cases.loc[:, '1/1/22':'12/31/22']
deaths = death_cases.loc[:, '1/1/22':'12/31/22']

dates = confirmed.keys()
world_cases = []
total_deaths = []

for i in dates:
    confirmed_sum = confirmed[i].sum()
    death_sum = deaths[i].sum()
    world_cases.append(confirmed_sum)
    total_deaths.append(death_sum)

days_since = np.array([i for i in range(len(dates))]).reshape(-1,1)
world_cases = np.array(world_cases).reshape(-1,1)
total_deaths = np.array(total_deaths).reshape(-1,1)
```

```python
world_cases
```

<font color = "red">Describe que es lo que sucede en el bloque de código anterior</font>

1. Se esta definiendo la información con la cual vamos a trabajar. Para *confirmed* y *deaths* se están selecioanndo las columnas que ven del 1/1/2022 al 31/02/22... todo el año 2022. Mientras que *cols* hace referencia a los casos confirmados usando el método .keys() se usa para recuperar todas las claves del diccionario, revisa todas las que están disponibles y arrojará una lista de todas las claves vigentes dentro del diccionario. Además *dates* recupera las fechas. Finalmente se crean dos listas vacias para el total de casos (*world_cases*) y defunciones (*total_deaths*).

2. Con el ciclo for se realiza la suma de los *confirmed* (i=indice) y de *deaths* y me agrega la informacion en las listas vacias creadas previamente: *world_cases* y *total_deaths*.

3. Se está cambiando la informacion generada a una lista, ya que el reshape permite cambiar la forma de una matriz sin cambiar los datos... en este caso lo estaría cambiando a una sola columna. 

```python
days_in_future = 20
future_forecast = np.array([i for i in range(len(dates)+days_in_future)]).reshape(-1,1)
adjusted_dates = future_forecast[:-days_in_future]
```

```python
future_forecast
```

```python
start = '1/22/2020'
start_date = datetime.datetime.strptime(start, '%m/%d/%Y')     #El método strptime crea un objeto que corresponde a la información del tiempo, desde una cadena o un format code.
future_forecast_dates = []
for i in range(len(future_forecast)):
    future_forecast_dates.append((start_date + datetime.timedelta(days=i)).strftime('%m/%d/%Y'))   # representa una duración, la diferencia entre dos fechas u horas.
```

```python
future_forecast_dates
```

<font color = "red">Describe para que hacemos el código anterior</font>

1. Se define una ventana de 20 dias. El reshape permite cambiar la forma de una matriz sin cambiar los datos. Así *future forecasts* contiene todos los dias del año 2022 + los 20 dias de la ventana, para un total de 0 a 384 dias. Finalmente *adjusted_dates* le resta los 20 dias de la ventana.

2. En la segunda parte del código se inicia el periodo en 1/22/2020 y se le da formato creando *start_date*. Se crea una lista vacia para *future_forecast_dates*. Asi el cíclo *for* hace un recorrido por *len(future_forecast* para agregar la fecha formateada.

A continuación, se utilizará el método de train_test_split con los datos de entrada: "days_since" y "world_cases". Además, se asigna el porcentaje correspondiente al split y el estado aleatorio.

```python
#Completa = train_test_split(#Completa)

X_train_confirmed, X_test_confirmed, y_train_confirmed, y_test_confirmed = train_test_split(days_since, world_cases, test_size=0.15, random_state=123) 
```
